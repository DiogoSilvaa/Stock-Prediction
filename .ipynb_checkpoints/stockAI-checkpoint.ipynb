```python
#Import libraries
import pandas_datareader as web
import numpy as np
import random
import time
from sklearn.preprocessing import scale
from collections import deque
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, BatchNormalization
from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint
```

```python
#Determine sequence length, prediction length and index to predict

SEQ_LEN = 100 #Amount of days used to predict
PRED_LEN = 5 #Amount of days predicted
PRED_IDX = 'QQQ' #Index to be predicted
BATCH_SIZE = 64
EPOCHS = 10
NAME = f"{PRED_IDX}-{SEQ_LEN}-Sequence-{PRED_LEN}-Prediction-{int(time.time())}"
```

```python
#Functions 

def create_target(currentP, futureP):
  if float(futureP) > float(currentP):
    return 1 # Price increased, good
  else:
    return 0 # Price decreased, bad

def preprocessing(df):
  df.drop('future', 1) # Future column only used to calculate targets

  #Normalise data 
  for c in df.columns:
    if c != "target": # Target is already preprocessed
      df[c] = df[c].pct_change()
      df[c] = scale(df[c])
      df.dropna(inplace=True)
  df.dropna(inplace=True)

  prev_days = deque(maxlen=SEQ_LEN) #If list reaches maxlen, old one are popped
  sequences = []

  for x in df.values:
    prev_days.append([n for n in x[:-1]]) #Creates a list of arrays containing SEQ_LEN sequence of rows 
    if len(prev_days) == SEQ_LEN:
      sequences.append([np.array(prev_days), x[-1]]) #Sequence of data and its respective target
  
  random.shuffle(sequences)

  #Balance out dataset 
  buy = [] #Buy signal sequence
  sell = [] #Sell signal sequence
  
  for seq, target in sequences: #Append sequences to its corresponding array
    if target == 0:
      sell.append([seq,target])
    elif target == 1:
      buy.append([seq,target])

  #Shuffle arrays
  random.shuffle(buy) 
  random.shuffle(sell)

  #Make sure there's the same amount of buy and sell signals
  lower = min(len(buy), len(sell))
  buy = buy[:lower]
  sell = sell[:lower]
  sequences = buy+sell

  random.shuffle(sequences)

  x = []
  y = []

  #Split arrays into array X (containing features) and array Y (containing targets)
  for seq, target in sequences:
    x.append(seq)
    y.append(target)

  return np.array(x), np.array(y)
    



```

```python
#Import stock data
stocks = ['QQQ','DIA','SPY']
main_df = pd.DataFrame()
for index in stocks:
  df = web.DataReader(index, data_source = 'yahoo', start='2000-01-01', end='2021-01-01') #Import data from Yahoo Finance (starting from 01-01-2000 to 01-01-2021)
  
  df[f"150SMA_{index}"] = df.Close.rolling(150).mean()  # Add a column to include the last 150 days simple moving average
  df[f"dSMA_{index}"] = df['Close'] - df[f"150SMA_{index}"] # Add a column to include distance of closing price to 150 SMA

  df.rename(columns={"Close": f"close_{index}", "Volume": f"volume_{index}"}, inplace=True)
  df = df[[f"close_{index}",f"volume_{index}", f"dSMA_{index}"]]
  
  if len(main_df) == 0:
    main_df = df
  else:
    main_df = main_df.join(df)

main_df['future'] = main_df[f"close_{PRED_IDX}"].shift(-PRED_LEN) # Shifts downards the amount of days in advance predicted
main_df['target'] = list(map(create_target, main_df[f"close_{PRED_IDX}"], main_df['future'])) # Add target to dataframe, 1 if price is going up in the days predicted in advance,0 if not

main_df.dropna(inplace=True)
```

```python
#Stock indexes price and volume history
fig = plt.figure(figsize=(15,15))
info = ['NASDAQ-100','Dow Jones Industrial','S&P 500']
colours = ['r','g','b']

#First subplot
plt = fig.add_subplot(2,2,1)
plt.set_xlabel('Date')
plt.set_ylabel('Price')
plt.set_title('3 US Stock Indexes Price History', fontsize = 25)
for index, colour in zip(stocks, colours):
  plt.plot(main_df[f'close_{index}'], colour)
plt.legend(info)

#Second subplot
plt = fig.add_subplot(2,2,2)
plt.set_xlabel('Date')
plt.set_ylabel('Volume')
plt.set_title('NASDAQ-100 Volume History', fontsize = 25)
plt.plot(main_df[f'volume_QQQ'], 'r')

#Third subplot
plt = fig.add_subplot(2,2,3)
plt.set_xlabel('Date')
plt.set_ylabel('Volume')
plt.set_title('Dow Jones Industrial Volume History', fontsize = 25)
plt.plot(main_df[f'volume_DIA'], 'g')

#Fourth subplot
plt = fig.add_subplot(2,2,4)
plt.set_xlabel('Date')
plt.set_ylabel('Volume')
plt.set_title('S&P 500 Volume History', fontsize = 25)
plt.plot(main_df[f'volume_SPY'], 'b')

fig.show()
```

```python
#Split main_df into main_df and val_main_df (95-5 ratio)

dates = sorted(main_df.index.values)
LAST_5PCT = dates[-(int(0.05*len(dates)))] 

val_main_df = main_df[(main_df.index >= LAST_5PCT)]
main_df = main_df[(main_df.index < LAST_5PCT)]

train_x, train_y = preprocessing(main_df)
test_x, test_y = preprocessing(val_main_df)


print(f"Training data type: {train_x.dtype}",f"Training target data type: {train_y.dtype}")
print(f"Testing data type: {test_x.dtype}",f"Test target data type: {test_y.dtype}")
print(f"Training data: {len(train_x)}, Validation: {len(test_x)}" )
#print(f"Training Sells: {for n in train_x: if n}, Training Buys: {train_y.count(1)}")
#print(f"Testing Sells: {test_y.count(0)}, Testing Buys: {test_y.count(1)}")
```

```python
#Model
model = Sequential()
model.add(LSTM(128, input_shape=(train_x.shape[1:]), return_sequences=True))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(LSTM(128, input_shape=(train_x.shape[1:]), return_sequences=True))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(LSTM(128, input_shape=(train_x.shape[1:])))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(32, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(2, activation="softmax"))

opt = tf.keras.optimizers.Adam(lr=0.001, decay=1e-6)

model.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

tensorboard = TensorBoard(log_dir=f'logs/{NAME}')

checkpoint_filepath = "models/RNN_Final-{epoch:02d}-{val_accuracy}.hd5"
checkpoint = ModelCheckpoint(filepath=checkpoint_filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

history = model.fit(
    train_x, train_y,
    batch_size = BATCH_SIZE,
    epochs = EPOCHS,
    validation_data=(test_x, test_y),
    callbacks=[tensorboard, checkpoint]
)
```

```python
# Score model
score = model.evaluate(test_x, test_y, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
# Save model
model.save("models/{}".format(NAME))
```
